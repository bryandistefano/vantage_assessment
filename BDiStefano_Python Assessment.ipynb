**2. Create an additional column in the DataFrame called CostPerTransaction.**

```python
df['CostPerTransaction'] = df['NumberOfItemsPurchased'] * df['CostPerItem']
df.head()
```

**3. Create an additional column in the DataFrame called SalesPerTransaction.**

```python
df['SalesPerTransaction'] = df['NumberOfItemsPurchased'] * df['SellingPricePerItem']
df.head()
```

**4. Create another column in the DataFrame called ProfitPerTransaction.**

```python
df['ProfitPerTransaction'] = df['SalesPerTransaction'] - df['CostPerTransaction']
df.head()
```

**5. Round the ProfitPerTransaction column to the nearest whole number.**

```python
df = df.round({'ProfitPerTransaction': 0})
df['ProfitPerTransaction'] = df['ProfitPerTransaction'].astype(int)
df.head()
```

After rounding the column to the nearest whole number, I converted the values to integers to discard the unnecessary decimal point.

**6. Combine the Day, Month, and Year columns into a new Date column in the format of "day-month-year".**

```python
df['Date'] = df['Day'].astype(str) + '-' + df['Month'] + '-' + df['Year'].astype(str)
df.head()
```

**7. Change the ItemDescription column in the DataFrame to lowercase.**

```python
df['ItemDescription'] = df['ItemDescription'].str.lower()
df.head()
```

**8. The ClientKeywords column should be split into 3 columns : ClientAge, ClientType, and LengthofContract.**

```python
df[['ClientAge', 'ClientType', 'LengthofContract']] = df['ClientKeywords'].str.split(',', expand=True)
df['ClientAge'] = df['ClientAge'].replace(regex=r"['\[\]]", value='')
df['ClientType'] = df['ClientType'].replace(regex=r"['\[\]]", value='')
df['LengthofContract'] = df['LengthofContract'].replace(regex=r"['\[\]]", value='')
df.head()
```

**9. Now that it has been split, drop the ClientKeywords column.**

```python
df = df.drop('ClientKeywords', axis=1)
df.head()
```

**10. Export the DataFrame as a CSV.**

```python
df.to_csv('transaction_updated.csv', index=False)
```
